Q1. What is salary distribution?

```python
sns.histplot(df["Average_Salary"],bins = 20)
```

Q2.Are there salary outliers?

```python
sns.boxplot(df["Average_Salary"])
```

Q3.How risky are jobs?

```python
sns.histplot(df["Automation_Probability_2030"],palette="rainbow")
```

How many jobs belong to each risk category?

```python

sns.countplot(x = "Risk_Category",data = df,palette="rainbow")
```

Which jobs pay most?

```python
top_salary = df.groupby("Job_Title")["Average_Salary"].mean().sort_values(ascending=False)

top_salary.round(2).head(10)
```

```python
top_salary.head(10).plot(kind="bar")
```

Which jobs are safest from AI?

```python
df.groupby(
    "Job_Title"
)[
    "Automation_Probability_2030"
].mean()
```

Does education increase salary?

```python
education = df.groupby("Education_Level")["Average_Salary"].mean().round(2)
sns.barplot(
    x="Education_Level",
    y="Average_Salary",
    data=df
)
```

Does experience increase salary?

```python
sns.scatterplot(x = "Years_Experience",y = "Average_Salary",data = df,palette="rainbow")
```

```python
top20 = df.head(10)
corr = (
    top20.select_dtypes(
        include=np.number
    )
    .corr()
)
```
